1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:

A) Rank

B) Name

C) Artist

D) Upload date

E) Views

```python
import selenium
import pandas as pd
import time
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException, ElementNotInteractableException
import requests
import re
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import warnings
warnings.filterwarnings('ignore')
```

```python
driver=webdriver.Chrome("chromedriver.exe")
time.sleep(2)

url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
```

```python
list_wiki = {}
list_wiki['Rank'] = []
list_wiki['Name'] = []
list_wiki['Artist'] = []
list_wiki['Upload date'] = []
list_wiki['Views'] = []


for i in range(1,31):
    
    rank = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[' + str(i) +']/td[1]'.format(i))
    for d in rank:
        list_wiki['Rank'].append(d.text)

    name = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[' + str(i) +']/td[2]'.format(i))
    for n in name:
        list_wiki['Name'].append(n.text)
      
        
    artist = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[' + str(i) +']/td[3]'.format(i))
    for a in artist:
        list_wiki['Artist'].append(a.text)
   

    upload_date = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[' + str(i) +']/td[5]'.format(i))
    for u in upload_date:
        list_wiki['Upload date'].append(u.text)
        
    
    views = driver.find_elements_by_xpath('//*[@id="mw-content-text"]/div[1]/table[3]/tbody/tr[' + str(i) +']/td[4]'.format(i))
    for v in views:
        list_wiki['Views'].append(v.text)


        
fund_df = pd.DataFrame(list_wiki)
fund_df


   
```
