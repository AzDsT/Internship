```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
name=[]
mileage=[]
dealer_name=[]
rating=[]
reviews=[]
price=[]

for i in range(1,400):
    website='https://www.cars.com/shopping/results/?page=' + str(i) + '&page_size=20&list_price_max=&makes[]=&maximum_distance=20&models[]=&stock_type=used&zip='
    response = requests.get(website)
    soup=BeautifulSoup(response.content, 'html.parser')
    results=soup.find_all('div', {'class':'vehicle-card'})
    
    for result in results:
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')
        
        
    
        try:
            mileage.append(result.find('div', {'class':'mileage'}).get_text())
        except:
            mileage.append('n/a')
        
        
        
        
        try:
            dealer_name.append(result.find('div', {'class':'dealer-name'}).get_text().strip())
        except:
            dealer_name.append('n/a')
        
        
        
        try:
            rating.append(result.find('span', {'class':'sds-rating__count'}).get_text())
        except:
            rating.append('n/a')
        
        
        
        try:
            reviews.append(result.find('span', {'class':'sds-rating__link'}).get_text())
        except:
            reviews.append('n/a')
        
        
        try:
            price.append(result.find('span', {'class':'primary-price'}).get_text())
        except:
            price.append('n/a')
        
```

```python
cars_com = pd.DataFrame({'Name':name, 'Mileage':mileage, 'Dealer Name':dealer_name, 'Rating':rating, 
                        'Reviews':reviews, 'Price':price})
```

```python
cars_com['Reviews']=cars_com['Reviews'].apply(lambda x:x.strip('reviews)').strip('('))
```
